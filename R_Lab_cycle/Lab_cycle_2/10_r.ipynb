```r
# Load the dataset
data <- read.csv("train.csv")
# Check the structure of the dataset
str(data)
# Summary of the dataset
summary(data)
# Fit a simple linear regression model
model <- lm(SalePrice ~ LotArea, data = data)
# Summary of the model
summary(model)
# Plotting SalePrice vs LotArea with regression line
plot(data$LotArea, data$SalePrice, xlab = "LotArea", ylab = "SalePrice")
abline(model, col = "red")

```

```r
# Fit a simple linear regression model
model <- lm(SalePrice ~ LotArea, data = data)

# Get regression coefficients (slope and intercept)
coefficients <- coef(model)
slope <- coefficients["LotArea"]
intercept <- coefficients["(Intercept)"]

# Display the slope and intercept
print(paste("Slope (Coefficient for LotArea):", slope))
print(paste("Intercept:", intercept))
```

```r
# Fit a simple linear regression model
model <- lm(SalePrice ~ LotArea, data = data)

# Create scatter plot
plot(data$LotArea, data$SalePrice, xlab = "LotArea", ylab = "SalePrice", main = "Scatter Plot with Regression Line")

# Add regression line
abline(model, col = "red")
```
